## Intro
    
The gym chain Model Fitness is developing a customer interaction strategy based on analytical data.
One of the most common problems gyms and other services face is customer churn. For a gym, it makes sense to say a customer has left if they don't come for a month. Of course, it's possible they're in Cancun and will resume their visits when they return, but that's not a typical case. Usually, if a customer joins, comes a few times, then disappears, they're unlikely to come back.
In order to fight churn, Model Fitness has digitized a number of its customer profiles. Your task is to analyze them and come up with a customer retention strategy.
You should:
- Learn to predict the probability of churn (for the upcoming month) for each customer
- Draw up typical user portraits: select the most outstanding groups and describe their main features
- Analyze the factors that impact churn most
- Draw basic conclusions and develop recommendations on how to improve customer service:
    - Identify target groups
    - Suggest measures to cut churn
    - Describe any other patterns you see with respect to interaction with customers

### Feature engeneering and feature selection

<a id='feature_e_s'></a>

Since 'avg_class_frequency_total and 'avg_class_frequency_current_month' are strongly correlated a new feature will be created out of them. The difference between is assumed to be rellevant in predicting churn because it seems reasnobale that if customers take less calsses in the current month than usuall they might be thinking of leaving the gym. In contrast, if costomers take more calsses than usual it seems reasnobale that they are happy with the gym and would like to stay.  

```python
df['diff_avg_class_frequency']=df['avg_class_frequency_total']-df['avg_class_frequency_current_month']
df.head()
```

```python
from scipy.stats import pointbiserialr

#import scipy.stats as stats
status=0
from scipy.stats import pointbiserialr
for col1 in ['churn']:
    for col2 in [ 'avg_class_frequency_total', 'avg_class_frequency_current_month','diff_avg_class_frequency']: 
        rpb, p_value = pointbiserialr(df[col1], df[col2])
        if (p_value< 0.05) and (abs(rpb)>0.2) :
            print("Point-biserialr correlation between", col1, "and", col2, "is significant:") 
            print("rpb=","{:.4f}".format(rpb), "p=","{:.4f}".format(p_value)) 
            print("")
            status+=1

if status<1:
    print("There is no sagnificant correlation that its effect size is greater than 0.2 ")
```

Point-biserialr correlation between churn and diff_avg_class_frequencyis significant, moderate and positive. As the difference between total class frequency and current frequency is higher (when total class frequency is higher than the current), a costomer is more likley to churn. 

Since the correlation between the total and current frequency is very strong and the correlation between `churn` and `total` is lower than the corellations between `churn` and `difference` or `current` variables, the `total` variable can be discarded to reduce multicollinearity. 

Using the `difference` and the `current` variables togther may be fine since they are only moderatly correlated (as seen below).

```python
import pandas as pd
#!pip install statsmodels
import statsmodels.formula.api as smf

log_reg_contract_continuos = smf.logit(formula="churn ~ contract_period", data=df).fit()

print(log_reg_contract_continuos.summary())

print(f'LLR P-value: = {log_reg_contract_continuos.llr_pvalue:.4f}')

print("------------------------------------------------------------------------------------------------------------")
print()
log_reg_contract_ordinal1 = smf.logit(formula="churn ~ C(contract_period, Treatment(reference=1))", data=df).fit()

print(log_reg_contract_ordinal1.summary())


print(f'LLR P-value: = {log_reg_contract_ordinal1.llr_pvalue:.4f}') # Regression p-value
print("------------------------------------------------------------------------------------------------------------")
print()
log_reg_contract_ordinal2 = smf.logit(formula="churn ~ C(contract_period, Treatment(reference=12))", data=df).fit()

print(log_reg_contract_ordinal2.summary())


print(f'LLR P-value: = {log_reg_contract_ordinal2.llr_pvalue:.4f}')
print("------------------------------------------------------------------------------------------------------------")
print()

log_reg_contract_ordinal3 = smf.logit(formula="churn ~ C(contract_period, Treatment(reference=6))", data=df).fit()

print(log_reg_contract_ordinal3.summary())


print(f'LLR P-value: = {log_reg_contract_ordinal3.llr_pvalue:.4f}')
print("------------------------------------------------------------------------------------------------------------")
print()

log_reg_month_end_continuos = smf.logit(formula="churn ~ month_to_end_contract", data=df).fit()

print(log_reg_month_end_continuos.summary())


print(f'LLR P-value: = {log_reg_month_end_continuos.llr_pvalue:.4f}')
print("------------------------------------------------------------------------------------------------------------")
print()
log_reg_month_end_ordinal1 = smf.logit(formula="churn ~ C(month_to_end_contract, Treatment(reference=1))", data=df).fit()

print(log_reg_month_end_ordinal1.summary())

print(f'LLR P-value: = {log_reg_month_end_ordinal1.llr_pvalue:.4f}')
print("------------------------------------------------------------------------------------------------------------")
print()
log_reg_month_end_ordinal2 = smf.logit(formula="churn ~ C(month_to_end_contract, Treatment(reference=12))", data=df).fit()

print(log_reg_month_end_ordinal2.summary())


print(f'LLR P-value: = {log_reg_month_end_ordinal2.llr_pvalue:.4f}')

```

as seen above all the binary logistic regression models are sigfnificant but the models with the highest Pseudo R-squre are those with contract_period as an ordinal variable. This suggets that since contract_period and month_to_end_contarct are strongly correlated, the latter can be discarded from the model to reduce multicollinearity. 

However, it may be worth while to check both optios if month_to_end_contract tells us whether the customer is **able to** churn in the next month. This idea will be investigated below:

```python
df.loc[(df['churn']==1) &(df['month_to_end_contract']==1) ,'month_to_end_contract'].count()
df.loc[(df['churn']==1) &(df['month_to_end_contract']==1) &(df['avg_class_frequency_current_month']>0) ,'month_to_end_contract'].count()

df.loc[(df['churn']==1) &(df['month_to_end_contract']>1) ,'month_to_end_contract'].count()
df.loc[(df['churn']==1) &(df['month_to_end_contract']>1) &(df['avg_class_frequency_current_month']>0) ,'month_to_end_contract'].count()
df.loc[(df['churn']==1) &(df['month_to_end_contract']>1) &(df['avg_class_frequency_current_month']>1) ,'month_to_end_contract'].count()
```

As seen above, customers can demand to end their contract and leave even before the end of contract, and even if they were active in the gym in the current month, probably as long as they are willing to pay a cancellation fee. So the month_to_end_contract variable does not necessarily give us an additional information over the contract_period variable. so month_to_end_contract can be discarded with a clear conscience.

```python
# load statmodels functions
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

# computing the vif for all given features
def compute_vif(df,considered_features): # adding a constant (intercept)
    
    x = add_constant(df[considered_features])
    # the calculation of variance inflation requires a constant
    
    
    # creating a dataframe to store vif values
    vif = pd.DataFrame()
    vif["Variable"] = x.columns
    vif["VIF"] = [variance_inflation_factor(x.values, i) for i in range(x.shape[1])]
    vif = vif[vif['Variable']!='const']
    return vif
```

#### Dealing with outliers by using winsorization
Creating different training sets to see whether outliers impact the models  

An additional training set will be created to see whether outliers impact the models. The outliers will be winsorized. Later on, results with and without winsorization will be compared.    

```python

# saving outliers percentage
out_percent_total = 0

# saving IQR thresholds of the training set
iqr_thresholds_dict = {} 


''' 
Winsorization function to replace outliers with the nearest values based on interquartile range and the traing set
in each feature 
''' 

def winsorization(data,data_train_winsorized,data_test_winsorized, parameter):
    
    global iqr_thresholds_dict
    global out_percent_total
    
    print('\nThe distribution of: {}'.format(parameter))
   
    sns.boxplot(data=data[parameter], palette="Set3",showfliers = True, orient="h")
    
    plt.show()
    if data[parameter].mean()>data[parameter].median():
        print('The distribution is skewed to the right')
    else:
        print('The distribution is skewed to the left')
    print('---------------------------------------')
    print('Statistics on: {}'.format(parameter))
    print(data[parameter].describe())
    print('\nThe median is:',data[parameter].median(),'\n')
    print('Variance is:', np.var(data[parameter]),'\n')
    mean=data[parameter].mean()
    std=data[parameter].std()
    var=data[parameter].var()
    cv=std/mean
    print('The coefficient of variation is:', cv,'\n') 
   
    print('---------------------------------------')
    Q1 = data[parameter].quantile(0.25)
    Q3 = data[parameter].quantile(0.75)
    IQR = Q3 - Q1
    min_iqr=Q1-(1.5*IQR)
    max_iqr=Q3+(1.5*IQR)
    print('---------------------------------------')
    print('min border:',min_iqr)
    print('max border:',max_iqr)
    print('---------------------------------------')
    
    # saving IQR thresholds of each feature in a nested dictionary (dictionary of dictionaries)
    
    iqr_thresholds_dict[parameter]= {'min': min_iqr, 'max': max_iqr}
      
    # Winsorizing outliers
    
    if (min_iqr<0) and (~(data[parameter]<0).any()):
        print('The outliers are considered to be values above',max_iqr)
        n_outliers=len(data[data[parameter]>max_iqr])
        print('We have',n_outliers,"values that we can consider outliers")
        if n_outliers>0:
            out_percent_temp = n_outliers/(data.size)
            out_percent_col=n_outliers/len(data[parameter])
            print("They are","{:.2%}".format(out_percent_col)," of",parameter, "and", "{:.2%}".format(out_percent_temp), " of the whole data set")
            out_percent_total += out_percent_temp
            
            data_train_winsorized.loc[data_train_winsorized[parameter]>max_iqr,parameter]=max_iqr
            data_test_winsorized.loc[data_test_winsorized[parameter]>max_iqr,parameter]=max_iqr
           
          
    else:
        print('The outliers are considered to be values below',min_iqr, "and above",max_iqr)
        n_outliers=len(data[(data[parameter]<min_iqr)|(data[parameter]>max_iqr)])
        print('We have',n_outliers,"values that we can consider outliers")
        if n_outliers>0:
            out_percent_temp = n_outliers/(data.size)
            out_percent_col=n_outliers/len(data[parameter])
            print("They are","{:.2%}".format(out_percent_col)," of",parameter, "and", "{:.2%}".format(out_percent_temp), " of the whole data set")
            out_percent_total+=out_percent_temp
   
            data_train_winsorized.loc[data_train_winsorized[parameter]>max_iqr,parameter]=max_iqr
            data_train_winsorized.loc[data_train_winsorized[parameter]<min_iqr,parameter]=min_iqr 
    
            data_test_winsorized.loc[data_test_winsorized[parameter]>max_iqr,parameter]=max_iqr
            data_test_winsorized.loc[data_test_winsorized[parameter]<min_iqr,parameter]=min_iqr 
    
   
                
# applying function

# creating new training and testing sets
x_train_IQR_winsorized=x_train_e_s.copy(deep=True)
x_test_IQR_winsorized=x_test_e_s.copy(deep=True)

for feature in ['age','avg_additional_charges_total', 'lifetime', 'diff_avg_class_frequency','avg_class_frequency_current_month']:
    winsorization(x_train_e_s,x_train_IQR_winsorized,x_test_IQR_winsorized,feature)
print('---------------------------------------')
print('---------------------------------------')
print('---------------------------------------')
print()
print("Percentage of all the outliers out of the whole training data set before winsorization is:","{:.2%}".format(out_percent_total))
```

#### Creating a pipline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, mutual_info_classif
from sklearn.linear_model import LogisticRegression

def get_pipeline():
    #  creating pipeline framework
    pipe = Pipeline([('scaler', StandardScaler()), 
                     ('selector', SelectKBest()), 
                     ('classifier', LogisticRegression())]) 
    return pipe
```

#### Hyper-parameters tuning with GridSearchCV

```python
    
def get_search_space(classifier_search):
    
    #creating search space
    search_space = [{'selector__k':[5,6,7,8,9,10,11],
                     'selector__score_func':[mutual_info_classif]}, 
                    classifier_search]
                    
    return search_space
```

```python

def get_classifier_search(name):
    if name == "LR":
        classifier_search= {'classifier': [LogisticRegression(random_state=0)],
                            'classifier__class_weight':['balanced', None],
                            'classifier__solver': ['lbfgs', 'newton-cg', 'newton-cholesky', 'sag', 'saga'], 
                            'classifier__penalty': [None],
                            'classifier__C': [0.01, 0.1, 1.0, 10, 100] 
                           }
        return classifier_search
    
    if name == "LR1":
        classifier_search= {'classifier': [LogisticRegression(random_state=0)],
                            'classifier__class_weight':['balanced', None],
                            'classifier__solver': ['liblinear', 'saga'],  
                            'classifier__penalty': ['l1'],
                            'classifier__C': [0.01, 0.1, 1.0, 10, 100]
                            }
              
        return classifier_search
        
    if name == "LR2":
        classifier_search= {'classifier': [LogisticRegression(random_state=0)],
                             'classifier__class_weight':['balanced', None],
                             'classifier__solver': ['lbfgs','liblinear','newton-cg','sag','newton-cholesky','saga'],  
                             'classifier__penalty': ['l2'],
                             'classifier__C': [0.01, 0.1, 1.0, 10, 100]
                            }
        return classifier_search
    
    elif name == "RF":
        classifier_search= {'classifier': [RandomForestClassifier(random_state=0)],
                            'classifier__n_estimators': [50, 100, 500], #The number of trees in the forest
                            'classifier__max_depth': [3, 5, 10, None], # The maximum depth of the tree
                             'classifier__max_features': ['sqrt', None], # The number of features to consider when looking for the best split, If None, then max_features=n_features,If “sqrt”, then max_features=sqrt(n_features),If “log2”, then max_features=log2(n_features)
                             'classifier__class_weight':['balanced','balanced_subsample', None]
                            }
        return classifier_search 

```

##### Nested cross validation

To prevent over fitting nested cross validation will be used:

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV 

model_names= ['LR','LR1','LR2','RF']

#for saving metrics
metrics_outer_cv=[]

#for saving grids and best estimators
grids_dict = {} 
best_estimatores={}

for model_name in model_names:

    
    list_of_tupes = [(x_train_e_s,y_train), 
                     (x_train_IQR_winsorized,y_train)] 
    
    for data_tuple in list_of_tupes:
        x_train_i, y_train_i = data_tuple
        
        pipe = get_pipeline()
        
        search_space = get_search_space(get_classifier_search(model_name))
        
        # Running the GridSearch
        
        inner_cv = RepeatedStratifiedKFold(n_splits=3,n_repeats=2, random_state=0) 
        
        outer_cv = RepeatedStratifiedKFold(n_splits=5,n_repeats=2, random_state=0)
        
        metrics_names = ['recall', 'precision','f1','accuracy','roc_auc', 'neg_log_loss'] 
        
        grid = GridSearchCV(estimator=pipe, param_grid=search_space, cv=inner_cv,  
                       scoring=metrics_names,refit='recall',return_train_score=False,n_jobs=-1,verbose=0) 
        #saving grids
        current_data_name=f'{model_name}_{x_train_i.name}'
    
        grids_dict[current_data_name] = grid
          
        # Running outer cross validation
            
        cv_model = cross_validate(grid, X=x_train_i, y=y_train_i, cv=outer_cv, scoring=metrics_names,
                                    return_estimator=False,return_train_score=False,n_jobs=-1,verbose=0)
            
             
          
        mean_accuracy=cv_model['test_accuracy'].mean() 
        mean_recall=cv_model['test_recall'].mean()
        mean_precision=cv_model['test_precision'].mean()
        mean_f1=cv_model['test_f1'].mean()
        mean_roc_auc=cv_model['test_roc_auc'].mean()
        mean_neg_log_loss=cv_model['test_neg_log_loss'].mean()
            
            
        # saving all metrics in a table
        metrics_outer_cv.append(classification_cv_metrics_to_list(current_data_name, mean_accuracy,
                                                                mean_recall, mean_precision,mean_f1,
                                                                mean_roc_auc,mean_neg_log_loss, 
                                                                ))
       
```

#### Silhouette score

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings("ignore")
from sklearn.metrics import silhouette_score


# a function that calculates the silhouette_score for each parameter and returns the highest silhouette_score for each parameter
def k_solution(data):
    score={}
    for n_cluster in [2,3,4,5,6,7]:
        kmeans = KMeans(n_clusters=n_cluster,random_state=0).fit(data)
    
        silhouette_avg = silhouette_score(data, kmeans.labels_)
        score[n_cluster]=silhouette_avg
    needed_number=max(score,key=score.get)
    return needed_number
```

```python
# applying the function for getting back the optimal number of clusters for the df
for data in [df_for_cluster_st,df_for_cluster_boxcox_st]:
    print(k_solution(data))
    
```

As seen above, for the only standardized data the highest silhouette_score was achieved by clustering to 5 groups. For the box-cox transformed and standardized data, the highest silhouette_score was achieved by clustering to 3 groups.
